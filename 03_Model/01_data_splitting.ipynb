# Phase 4: Data Splitting
In this step, we divide our scaled dataset into training and testing sets. 
* **Training Set (80%)**: Used to teach the model.
* **Testing Set (20%)**: Used to evaluate the model's performance on unseen data.
Saving these as separate files ensures consistency between our training and evaluation notebooks.

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Load the scaled dataset
file_path = '../01_Data/04_GlobalLandTemperatures_scaled.csv'
df = pd.read_csv(file_path)

# 2. Perform the 80/20 split
# We use random_state=42 to ensure the split is reproducible
train_df, test_df = train_test_split(df, test_size=0.20, random_state=42)

# 3. Save the split files
train_df.to_csv('../01_Data/train_data.csv', index=False)
test_df.to_csv('../01_Data/test_data.csv', index=False)

print(f"Data split successfully!")
print(f"Training samples: {len(train_df)}")
print(f"Testing samples: {len(test_df)}")
```
