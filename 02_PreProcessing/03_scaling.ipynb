# Phase 3: Data Scaling and Normalization
In this notebook, we prepare our numerical features for the neural network. 
Key objectives include:
* **Feature Scaling**: Using `MinMaxScaler` to bring all input features into a range of [0, 1].
* **Target Scaling**: Normalizing the `AverageTemperature` to ensure stable gradients during training.
* **Data Persistence**: Saving the scaled arrays and the scalers themselves for model training and future evaluation.

### Step 1: Environment Setup and Data Loading
We import the necessary libraries, including `joblib` for saving our scalers, and load the engineered dataset created in the previous phase.

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import  MinMaxScaler
# Load the dataset
file_path = '../01_Data/03_GlobalLandTemperatures_encoded.csv'
df = pd.read_csv(file_path)
```

### Step 2: Applying MinMaxScaler
Neural networks perform best when input features are on a similar scale. We apply the `MinMaxScaler` to:
1. **Features (X)**: `Year`, `Month`, `Lat_Num`, `Lon_Num`, and our encoded location labels.
2. **Target (y)**: The `AverageTemperature`. 

**Note:** We save the `scaler_y` separately so we can reverse the scaling later to see the temperatures in degrees Celsius.

```python
features = ['Year', 'Month', 'Lat_Num', 'Lon_Num', 'City_Encoded', 'Country_Encoded']
X = df[features]
y = df['AverageTemperature']
    
scaler_X, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).flatten()
```

### Step 3: Reconstructing the Dataset
We convert the resulting NumPy arrays back into a pandas DataFrame. This allows us to easily verify the scaling and save a human-readable version of the processed data to a CSV file.

```python
# 2. Convert scaled arrays back into a DataFrame
# We use the same column names from our 'features' list
df_scaled = pd.DataFrame(X_scaled, columns=features)

# 3. Add the scaled target back
df_scaled['AverageTemperature_Scaled'] = y_scaled
# Save the updated data to a new CSV file
df_scaled.to_csv('../01_Data/04_GlobalLandTemperatures_scaled.csv', index=False)
print('Updated file saved successfully.')
```

### Step 4: Saving Processed Arrays and Scaler Objects
To streamline the model training phase, we save:
* **NumPy Arrays**: `X_scaled` and `y_scaled` for fast loading.
* **Scalers**: `scaler_X.pkl` and `scaler_y.pkl`. These are critical for processing new data or evaluating the model in the future.

```python

# 1. Save the processed NumPy arrays
np.save('../01_Data/X_scaled.npy', X_scaled)
np.save('../01_Data/y_scaled.npy', y_scaled)

# 2. Save the scalers and encoders
joblib.dump(scaler_X, '../01_Data/scaler_X.pkl')
joblib.dump(scaler_y, '../01_Data/scaler_y.pkl')

print("All files saved successfully!")
```
